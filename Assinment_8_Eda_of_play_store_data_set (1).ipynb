```python
# checking  columns
for col in df.columns:
    col1 = col.replace(' ','')
    df = df.rename(columns={col:col1})
df.columns
```

```python
#Dropping the following columns - AppId, DeveloperWebsite, DeveloperEmail, PrivacyPolicy, Currency, DeveloperId, ScrapedTime, MinimumAndroid that seemd not relevent
df = df.drop(['AppId','DeveloperWebsite','DeveloperEmail','PrivacyPolicy','Currency','DeveloperId','ScrapedTime','MinimumAndroid'],axis=1)
df.head()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Percentage of null values
df.isnull().sum() / len(df) * 100
```

```python
#checking null values
df.isnull().sum()
```

```python
#removing or replacing null values
#As we see most importent data is rating ,rating counts, installs, so we  replace these data with mean  and remove other

mean = df['Rating'].mean()
mean
```

```python
#replacing mean in place of null values
df["Rating"].fillna(mean,inplace=True)
```

```python
null_count = df['Rating'].isnull().sum()

print(f"Number of null values in the 'Rating' column: {null_count}")
```

```python
#replacing rating counts with its mean
Rating_Count_mean = df["RatingCount"].mean()
Rating_Count_mean
```

```python
#replacing mean in place of null values
df["RatingCount"].fillna(Rating_Count_mean,inplace=True)
```

```python
# Dropping all the rows having nan values in the given columns
df = df.dropna()
```

```python
df.isnull().sum()
```

# Category


```python
df['Category'].value_counts()
```

```python
# Make Music catagor 1 bcz its label with different name but in real its lie under 1  'Music' and 'Music & Audio' into 'Music'
df['Category'] = df['Category'].str.replace('Music & Audio', 'Music')
```

```python
# Get the count max in catagory
max_count = df['Category'].value_counts().max()

# Get the category name(s) corresponding to the maximum count
max_count_categories = df['Category'].value_counts().loc[lambda x: x == max_count].index.tolist()

print(max_count_categories,max_count)
```

```python
# Get the minimumcount of a category
min_count = df['Category'].value_counts().min()

# Get the category name(s) corresponding to the minimum count
min_count_categories = df['Category'].value_counts().loc[lambda x: x == min_count].index.tolist()

print(min_count_categories,min_count)
```

# *Rating*

```python
# Get the minimumcount of a Rating
min_count = df['Rating'].value_counts().min()

# Get the Rating name(s) corresponding to the minimum count
min_count_Rating = df['Rating'].value_counts().loc[lambda x: x == min_count].index.tolist()

print(min_count_Rating,min_count)
```

```python
# Get the maximumcount of a Rating
max_count = df['Rating'].value_counts().max()

# Get the Rating name(s) corresponding to the minimum count
max_count_Rating = df['Rating'].value_counts().loc[lambda x: x == max_count].index.tolist()

print(max_count_Rating,max_count)
```

```python
df['ContentRating']=df['ContentRating'].replace('Unrated',"Everyone")

#Cleaning other values just to include Everyone, Teens and Adult
df['ContentRating']=df['ContentRating'].replace('Mature 17+',"Adults")
df['ContentRating']=df['ContentRating'].replace('Adults only 18+',"Adults")
df['ContentRating']=df['ContentRating'].replace('Everyone 10+',"Everyone")
df['ContentRating']
```

```python
#checking dtype of Installs
t=df['Installs'].dtype
t
```

```python
#handling non-numeric values
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

```

```python
df['Installs'].fillna(0, inplace=True)

# Convert the 'Installs' column to integer
df['Installs'] = df['Installs'].astype(int)



```

```python
#Minumum install
df['MinimumInstalls'].fillna(0, inplace=True)

# Convert the 'Installs' column to integer
df['MinimumInstalls'] = df['MinimumInstalls'].astype(int)
```

```python
df.dtypes
```

```python
df.head(10)
```

```python
# Add a new column 'RatingType' with default value 'NoRating'
df['RatingType'] = 'NoRating'

# Categorize based on the provided bins and labels
df.loc[(df['RatingCount'] > 0) & (df['RatingCount'] <= 10000.0), 'RatingType'] = 'Less than 10K'
df.loc[(df['RatingCount'] > 10000) & (df['RatingCount'] <= 500000.0), 'RatingType'] = 'Between 10K and 500K'
df.loc[(df['RatingCount'] > 500000), 'RatingType'] = 'More than 500K'

# Display the counts of each RatingType
rating_type_counts = df['RatingType'].value_counts()
print(rating_type_counts)
```

# Data Visualization

```python
df.columns
```

```python

corr_matrix = df.corr()
# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
# Set the title
plt.title('Correlation Matrix Heatmap')
# Display the chart
plt.show()
```

```python
#Distribution of Apps by Category
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Category', order=df['Category'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribution of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#Distribution of App Ratings
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Rating', bins=4, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Rating vs. Rating Count
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='RatingCount', y='Rating')
plt.title('Rating vs. Rating Count')
plt.xlabel('Rating Count')
plt.ylabel('Rating')
plt.xscale('log')
plt.tight_layout()
plt.show()
```

```python
#Rating Distribution by Category
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Category', y='Rating')
plt.xticks(rotation=90)
plt.title('Rating Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
#istribution of Apps by Rating Type
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='RatingType', order=df['RatingType'].value_counts().index)
plt.title('Distribution of Apps by Rating Type')
plt.xlabel('Rating Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='RatingType', order=df['RatingType'].value_counts().index)
plt.title('Distribution of Apps by Rating Type')
plt.xlabel('Rating Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#Distribution of Apps by Rating Type'
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='RatingType', order=df['RatingType'].value_counts().index)
plt.title('Distribution of Apps by Rating Type')
plt.xlabel('Rating Type')
plt.ylabel('counts')
plt.tight_layout()
plt.show()
```

```python
#App Releases Over Time
df['Released'] = pd.to_datetime(df['Released'])
df['Year'] = df['Released'].dt.year

plt.figure(figsize=(10, 6))
sns.lineplot(data=df.groupby('Year')['AppName'].count().reset_index(), x='Year', y='AppName')
plt.title('App Releases Over Time')
plt.xlabel('Year')
plt.ylabel('Number of App Releases')
plt.tight_layout()
plt.show()
```

```python
#Distribution of Apps by Content Rating
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='ContentRating', order=df['ContentRating'].value_counts().index)
plt.title('Distribution of Apps by Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
#Rating Distribution by Content Rating
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='ContentRating', y='Rating', order=df.groupby('ContentRating')['Rating'].median().sort_values().index)
plt.title('Rating Distribution by Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Distribution of Rating Types by Content Rating
rating_type_content = df.groupby(['ContentRating', 'RatingType']).size().unstack().fillna(0)
rating_type_content.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Rating Types by Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Average Rating of Apps Over Time
df['Year'] = df['Released'].dt.year
avg_rating_by_year = df.groupby('Year')['Rating'].mean()

plt.figure(figsize=(10, 6))
plt.plot(avg_rating_by_year.index, avg_rating_by_year.values, marker='o')
plt.title('Average Rating of Apps Over Time')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
#Rating Types Over Time'
rating_type_over_time = df.groupby(['Year', 'RatingType']).size().unstack().fillna(0)

plt.figure(figsize=(10, 6))
rating_type_over_time.plot(kind='area', stacked=True)
plt.title('Rating Types Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#In-App Purchases and Ad Support by Rating Type
inapp_ads_by_rating = df.groupby('RatingType')[['InAppPurchases', 'AdSupported']].sum()

plt.figure(figsize=(10, 6))
inapp_ads_by_rating.plot(kind='bar')
plt.title('In-App Purchases and Ad Support by Rating Type')
plt.xlabel('Rating Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
